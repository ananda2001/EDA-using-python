```python
import pandas as pd

df = pd.read_csv('tmdb-movies.csv')
```

```python
df.info()
df.describe()
```

```python
# converting the data type of release_date to datetime

df['release_date'] = pd.to_datetime(df['release_date'])

# removing duplicate rows, if any

df.drop_duplicates(inplace = True)

# removing unessential columns as stated below

df.drop(['budget_adj','revenue_adj','overview','id','imdb_id','homepage','tagline'],axis =1, inplace=True)
```

```python
# removing all rows with null entries anywhere

df = df.dropna(how='any',axis=0)

df.info()
```

```python
# removing all rows with budget = 0

df_nonzero_budget = df[df['budget'] > 0]

#removing all rows with revenue = 0 and thus obtaining a cleaned dataframe

clean_df = df_nonzero_budget[df_nonzero_budget['revenue'] > 0]

clean_df.info()
clean_df.describe()
```

```python
#sorting all movies by budget in ascending order

ascend_budget_df = clean_df.sort_values(by='budget', ascending=True)

# displaying the 5 movies with lowest budgets

ascend_budget_df.head(5)
```

```python
# displaying the 5 movies with highest budgets

ascend_budget_df.tail(5)
```

```python
# creating a dataframe containing all movies released in or after 2000

movies_after_2000_df = clean_df[clean_df['release_year'] >= 2000]

# creating a dataframe containing all movies released between 2000 and 2005

movies_between_2000_and_2005_df = movies_after_2000_df[movies_after_2000_df['release_year'] <= 2005]

# storing all the number of words in each movie title of the above dataframe in a separate dataframe

titles_df = movies_between_2000_and_2005_df['original_title'].str.split().str.len()

titles_df
```

```python
# calculating average number of words in movie titles between the years 2000 and 2005 and then displaying the result

av_no_of_words = titles_df.mean(axis=0)
print (av_no_of_words)
```

```python
#sorting all movies by revenue in ascending order

ascend_revenue_df = clean_df.sort_values(by='revenue', ascending=True)

# display the 2 movies with highest revenue

ascend_revenue_df.tail(2)
```

```python
# displaying the 2 movies with lowest revenue

ascend_revenue_df.head(2)
```

```python
# creating a dataframe containing all movies released in year 2006

release_year_2006_df = clean_df[clean_df['release_year'] == 2006]

# creating another dataframe containing only the runtime column

runtime_2006_df = pd.DataFrame(release_year_2006_df,columns=['runtime'])

runtime_2006_df
```

```python
#calculating average runtime of movies in the year 2006 and then displaying the result

av_runtime_2006 = runtime_2006_df.mean(axis=0)
print (av_runtime_2006)
```
